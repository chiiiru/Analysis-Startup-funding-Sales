# **Analysis of startup funding and valuation trends**

Inspecting the dataset to see what the fields look like.

```python
--Inspect the data and limit the output to 5 records
select * from public.companies
limit 5 
```

```python
--Inspect the data and limit the output to 5 records
select * from funding 
limit 5
```

1.  Which countries and continents have the highest total funding for startup?

```python
SELECT DISTINCT public.companies.country, public.companies.continent, COUNT(funding) AS "number of funding"
FROM public.companies
INNER JOIN public.funding ON public.companies.company_id = public.funding.company_id
GROUP BY public.companies.country, public.companies.continent
ORDER BY "number of funding" DESC
LIMIT 5;



```

2.  What are the top industries that receive the most funding and have the highest valuations?

```python
select count(Industry), industry, funding, valuation from public.funding
inner join public.industries
on public.funding.company_id = public.industries.company_id
group by public.funding.valuation, public.funding.funding, public.industries.industry
order by count desc
limit 5 
```

3.  How does the funding landscape differ between developed and emerging markets?


```python
SELECT
    CASE 
        WHEN continent = 'North America' OR continent = 'Europe' THEN 'Developed'
        ELSE 'Emerging' 
    END AS market,
    AVG(funding) AS avg_funding,
    COUNT(*) AS num_startups
FROM
    public.companies
    INNER JOIN public.funding
    ON public.companies.company_id = public.funding.company_id
GROUP BY 
    market
```

4.  What are the top 10 most funded startups and what industries are they in?


```python
select company, sum(funding) as "Funding", industry
from public.companies
inner join public.funding on public.companies.company_id = funding.company_id
inner join public.industries on companies.company_id = public.industries.company_id
group by company, industry
order by "Funding" desc
limit 10
```

5.  What are the top 5 investors by total funding and how do their investment strategies vary by industry?

```python
select select_investors as "Investor", sum(funding) as "Total_Funding",
industry
from public.funding 
inner join public.industries
on funding.company_id = industries.company_id
group by  industry, select_investors
order by "Total_Funding" desc 
limit 5
```

6.   Which industries have the highest growth rates in terms of funding and valuations?

```python
select industry, sum(funding) as "funding", sum(valuation) as "Valuation"
from funding
inner join public.industries
on funding.company_id = industries.company_id
group by industry
order by "funding" desc 
limit 5
```

# Sales analysis

1.  Which product lines have the highest sales and revenue?

```python
SELECT product_line, SUM(quantity) AS total_sales, SUM(total) AS total_revenue
FROM sales
GROUP BY product_line
ORDER BY total_revenue DESC
LIMIT 5;
```

2.  What is the average order size and revenue per order for each warehouse location?

```python
select warehouse, avg(quantity) as "average_order_size", sum(total)/sum(quantity) as "revenue_per_order"
from public.sales
group by warehouse
limit 10
```

3.  How does sales performance vary across different client types?

```python
select client_type, sum(quantity) as "Sales", sum(total) as "Amount"
from public.sales
group by client_type
```

4.  Which product line generates the highest profit margin?

```python
select product_line, (sum(total) - sum((quantity)+(unit_price))) as "Profit"
from public.sales
group by product_line
order by "Profit" desc
limit 5
```

5.   Which client type has the highest average order value?

```python
select client_type, avg(total) as "highest_average_order" from public.sales
group by client_type 
ORDER BY client_type desc


```
